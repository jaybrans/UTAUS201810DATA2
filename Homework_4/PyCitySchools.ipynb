```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data = pd.read_csv(school_data_to_load)
totalSchools = len(school_data.index)
totalStudents = len(student_data.index)
totalMathScore = sum(student_data['math_score'])
totalReadingScore = sum(student_data['reading_score'])
totalBudget = sum(school_data['budget'])
averageMathScore = totalMathScore/float(totalStudents)
averageReadingScore = totalReadingScore/float(totalStudents)
passingMathRows = school_data_complete.loc[school_data_complete['math_score']>69]
passingReadingRows = school_data_complete.loc[(school_data_complete['reading_score'])>69]
passingMath = float((len(passingMathRows)))/totalStudents
passingReading = float(len(passingReadingRows))/totalStudents
overallPassing = (passingMath+passingReading)/2

df = pd.DataFrame(columns=['Total Schools', 'Total Students', 'Total Math Score', 
                           'Total Reading Score', 'Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate'])
df.loc[df.index.max() + 1] = [totalSchools, totalStudents, totalMathScore, totalReadingScore, totalBudget, averageMathScore, averageReadingScore, passingMath*100, passingReading*100, overallPassing*100]
df.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
myDf = pd.DataFrame(columns=['School Name', 'School Type', 'Total Students',
                             'Total School Budget', 'Per Student Budget', 'Average Math Score',
                             'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate'])


for index, row in school_data.iterrows():
    schoolName = row['school_name']
    totalStudentRows = school_data_complete.loc[school_data_complete['school_name'] == schoolName]
    totalMathScore = sum(totalStudentRows['math_score'])
    totalReadingScore = sum(totalStudentRows['reading_score'])
    schoolBudget = int(row['budget'])
    totalStudents = len(totalStudentRows)
    perStudentBudget = schoolBudget/totalStudents
    avgMathScore = totalMathScore/len(totalStudentRows)
    avgReadingScore = totalReadingScore/len(totalStudentRows)
    passingMathRows = totalStudentRows.loc[totalStudentRows['math_score']>69]
    passingReadingRows = totalStudentRows.loc[(totalStudentRows['reading_score'])>69]
    passingMath = float((len(passingMathRows)))/totalStudents
    passingReading = float(len(passingReadingRows))/totalStudents
    overallPassing = (passingMath+passingReading)/2
    myDf.loc[index] = [schoolName, row['type'], totalStudents, schoolBudget, 
               perStudentBudget, avgMathScore, avgReadingScore, passingMath*100, passingReading*100, overallPassing*100]
    
myDf = myDf.sort_values(by=['Overall Passing Rate'], ascending = False)
myDf.head(5)
```

## Bottom Performing Schools (By Passing Rate)

```python
myDf = myDf.sort_values(by=['Overall Passing Rate'], ascending = True)
myDf.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.

```python
mathDf = pd.DataFrame(columns=['9th', '10th', '11th', '12th'])

for index, row in school_data.iterrows():
    nineRows = school_data_complete.loc[(school_data_complete['grade']=="9th") & (school_data_complete['school_name']==row['school_name'])]
    tenRows = school_data_complete.loc[(school_data_complete['grade']=="10th") & (school_data_complete['school_name']==row['school_name'])]
    elevenRows = school_data_complete.loc[(school_data_complete['grade']=="11th") & (school_data_complete['school_name']==row['school_name'])]
    twelthRows = school_data_complete.loc[(school_data_complete['grade']=="12th") & (school_data_complete['school_name']==row['school_name'])]
    nineAverage = sum(nineRows['math_score'])/float(len(nineRows))
    tenAverage = sum(tenRows['math_score'])/float(len(tenRows))
    elevenAverage = sum(elevenRows['math_score'])/float(len(elevenRows))
    twelthAverage = sum(twelthRows['math_score'])/float(len(twelthRows))
    mathDf.loc[row['school_name']] = [nineAverage, tenAverage, elevenAverage, twelthAverage]
mathDf.head(15)

```
